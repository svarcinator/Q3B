```python
logs_path = "./results/log_test.log_test.2024-04-26_19-53-33.logfiles/"
```

```python
def get_logs(adir):
    log_list = []
    log_re = re.compile("(.*.log)$")
    for root, dirs, files in os.walk(adir):
        for file in files:
            if log_re.match(file):
                
                # open file, read it and save its content to log list
                file_path = os.path.join(root, file)
                
                with open(file_path, 'r', encoding='utf-8') as log_file:
                    log_content = log_file.read()
                    
                    log_list.append(parse_log(log_content))
    return log_list
                
```

```python
def init_dict():
    adict = {"benchmark-family" : "",
             "name" : "",
             "data" : {
                 "time" : 0,
                      "bw" : 0,
                       "approx-type" : "",
                       "cache-hits" : 
                        {
                           "bvecExpr" : 0,
                            "preciseBvecs" : 0,
                            "sameBWPreciseBvecs" : 0,
                            "prevBWpreciseBvecs" : 0,
                            "intervals" : 0,
                            "bddExprCache" : 0,
                            "sameBWPreciseBdds" : 0,
                            "sameBWImpreciseBvecStates" : 0
                            
                            
            }
        }
    }
    return adict 
```

```python
def parse_command(log_content):
    command_re = re.compile(".*\/BV\/(.*)\/(.*smt2)")
    if command_re.match(log_content):
        match = command_re.match(log_content)
        
    return match.group(1), match.group(2)
    
```

```python
def parse_data(log_content):
    data_re = re.compile(".*(\[.*\}).*")
    match = data_re.match(log_content)
    print(match)
```

```python
def parse_log(log_content):
    parsed = init_dict()
    family, name = parse_command(log_content)
    parsed["benchmark-family"] = family
    parsed["name"] = name
    
    parse_data(log_content)
    
    
    return parsed 
```
